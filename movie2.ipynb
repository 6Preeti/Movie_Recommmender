```python
import numpy as np
import pandas as pd
```

```python
credits=pd.read_csv('credits.csv')
credits
```

```python
meta = pd.read_csv('movies_metadata.csv')
```

```python
meta['release_date'] = pd.to_datetime(meta['release_date'], errors='coerce')
```

```python
meta['year'] = meta['release_date'].dt.year
```

```python
meta['year'].value_counts().sort_index()
```

```python
new_meta = meta.loc[meta.year == 2017,['genres','id','title','year']]
```

```python
new_meta
```

```python
new_meta['id'] = new_meta['id'].astype(int)
```

```python
movies = pd.merge(new_meta, credits, on='id')
```

```python
movies
```

```python
# an expression node or a string containing a Python literal or container display
import ast
movies['genres'] = movies['genres'].map(lambda x: ast.literal_eval(x))
movies['cast'] = movies['cast'].map(lambda x: ast.literal_eval(x))
movies['crew'] = movies['crew'].map(lambda x: ast.literal_eval(x))
```

```python
def gen_List(x):
    L1 = []
    st = " "
    for i in x:
        if i.get('name') == 'Science Fiction':
            scifi = 'Sci-Fic'
            L1.append(scifi)
        else:
            L1.append(i.get('name'))
    if L1 == []:
        return np.NaN
    else:
        return (st.join(L1))
```

```python
movies['genres_list'] = movies['genres'].map(lambda x: gen_List(x))
movies['genres_list']
```

```python
def actor1(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if (casts == []):
        return np.NaN
    else:
        return (casts[0])
```

```python
#getting actor1 name
movies['actor_1_name'] = movies['cast'].map(lambda x: actor1(x))
```

```python
def actor2(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if (casts == [] or len(casts)<=1):
        return np.NaN
    else:
        return (casts[1])
```

```python
#getting actor2 name
movies['actor_2_name'] = movies['cast'].map(lambda x: actor2(x))
```

```python
def actor3(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if (casts == [] or len(casts)<=2):
        return np.NaN
    else:
        return (casts[2])
```

```python
#getting actor3 name
movies['actor_3_name'] = movies['cast'].map(lambda x: actor3(x))
```

```python
def directors(x):
    dirts = []
    s = " "
    for i in x:
        if i.get('job') == 'Director':
            dirts.append(i.get('name'))
    if dirts == []:
        return np.NaN
    else:
        return (s.join(dirts))
```

```python
movies['director_name'] = movies['crew'].map(lambda x: directors(x))
```

```python
casting = movies.loc[:,['director_name','actor_1_name','actor_2_name','actor_3_name','genres_list','title']]
```

```python
casting
```

```python
casting.isna().sum()
```

```python
casting = casting.dropna(how='any')
casting.isna().sum()
```

```python
casting = casting.rename(columns={'genres_list':'genres'})
casting = casting.rename(columns={'title':'movie_title'})
casting['movie_title'] = casting['movie_title'].str.lower()
casting['combine_all'] = casting['actor_1_name'] + ' ' + casting['actor_2_name'] + ' '+ casting['actor_3_name'] + ' '+ casting['director_name'] +' ' + casting['genres']
casting
```
